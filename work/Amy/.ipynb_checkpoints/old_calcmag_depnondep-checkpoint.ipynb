```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import math

import os
import sys
from pathlib import Path

# フォルダパスを指定してsys.pathに追加
sys.path.append(str(Path(os.getcwd()).resolve().parent / "func"))

# 関数が格納されているモジュールをインポート
import amalysis as amalysis
```

```python
raw_path = str(Path(os.getcwd()).resolve()) + '/data/Amy_dissimilarity.csv'
raw = pd.read_csv(raw_path, header=None).T
raw.index = ['word1'] + ['word2'] + list(range(1, raw.shape[0]-1))
raw = raw.rename(columns={0:'BDI'})
raw.loc['word2', 'BDI'] = np.nan
head = raw.iloc[:2, :].copy()

df_raw = raw.iloc[2:, :].copy()
df_raw = df_raw.sort_values('BDI', ignore_index=True)
df_raw = df_raw.drop('BDI', axis=1)
df_raw = df_raw # dissimilarity (make it similarity in exponential part)
df_raw = pd.concat([head, df_raw]).drop('BDI', axis=1).copy()
df_raw
```

```python
# Make high1/2, low1/2 groups which only have dissimilarity scores
df_low1 = pd.concat([head, df_raw[df_raw.index.isin(np.arange(0, 60, 2))]]).drop('BDI', axis=1).copy()
df_low2 = pd.concat([head, df_raw[df_raw.index.isin(np.arange(1, 60, 2))]]).drop('BDI', axis=1).copy()
df_high1  = pd.concat([head, df_raw[df_raw.index.isin(np.arange(60, 120, 2))]]).drop('BDI', axis=1).copy()
df_high2  = pd.concat([head, df_raw[df_raw.index.isin(np.arange(61, 120, 2))]]).drop('BDI', axis=1).copy()

df_60 = [df_low1, df_low2, df_high1, df_high2]
```

```python
# make a "similarity matrix" in tlab: word1 and 2 are inordered, use unique word in case you didn't order it
unique_words = np.array(['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'brown', 'grey', 'black', 'happiness', 'joy', 'calm', 'confidence', 'envy', 'fear', 'anxiety', 'disgust', 'sadness', 'confusion', 'boredom', 'defeated', 'anger'])
df_dissim = []

for idx, arr in enumerate(df_60):
    dissim_m_temp, _ = amalysis.create_dissimilarity_matrix(arr, unique_words)
    df_dissim.append(dissim_m_temp)
```

```python
for i in range(len(df_60)):
    amalysis.plot_dissimilarity_heatmap(df_dissim[i], annotation=False, mode_dissim=True, title=f"Dissimilarity matrix_{i}")
```

```python
# make exponential of similarity matrix to make it distance, "similarity matrix" in Leinster

df_dist = []

for mat in range(len(df_60)):
    df_temp = df_dissim[mat].copy()
    for i in range(len(df_temp)):
        for j in range(len(df_temp.iloc[0, :])):
            df_temp.iloc[i, j] = math.e ** (-1 * df_dissim[mat].iloc[i, j]) # dissimilarity -> similarity
    df_dist.append(df_temp)
```

```python
# visualize df_dist

for mat in range(len(df_dist)):
    amalysis.plot_dissimilarity_heatmap(df_dist[mat], annotation=False, mode_dissim=False, title=f"Similarity matrix (distance)_{mat}")
```

```python
# calculate inversed dist_mat (M) 

mat_dist_inv = []
df_dist_inv = []
for i in range(len(df_dist)):
    mat_dist = df_dist[i].to_numpy()
    mat_dist_inv.append(np.linalg.inv(mat_dist)) # mat_dist_inv @ mat_dist で丸め誤差以内（たぶん？うまくいかなきゃエラーでるぽいし）
    df_dist_inv.append(pd.DataFrame(mat_dist_inv[i], index=unique_words, columns=unique_words))
    amalysis.plot_dissimilarity_heatmap(df_dist_inv[i], annotation=False, mode_dissim=False, title=f"Inverse matrix of the similarity matrix_{i}")
```

```python
# calculate magnitude of M
magnitude = np.zeros(len(mat_dist_inv))
for i in range(len(mat_dist_inv)):
    magnitude[i] += np.sum(mat_dist_inv[i])
print(f'magnitude for all subject, color and emotion: {magnitude}')
```

```python
for i in range(len(mat_dist_inv)):
    if i<3:
        cond = 'high depressive'
    else:
        cond = 'low depressive'
    mat_dist = df_dist[i].to_numpy()
    mat_dist_col = mat_dist[0:10,0:10]
    mat_dist_emo = mat_dist[10:23, 10:23]
    
    mat_dist_col_inv = np.linalg.inv(mat_dist_col)
    mat_dist_emo_inv = np.linalg.inv(mat_dist_emo)
    
    mag_col = np.sum(mat_dist_col_inv)
    mag_emo = np.sum(mat_dist_emo_inv)
    
    print(f'condition: {cond}')
    print(f'Magnitude for the subject-averaged data, includes color and emotion: {magnitude[i]:.3f}')
    print(f'Magnitude for the subject-averaged data, includes only color:        {mag_col:.3f}')
    print(f'Magnitude for the subject-averaged data, includes only emotion:      {mag_emo:.3f}')
```
